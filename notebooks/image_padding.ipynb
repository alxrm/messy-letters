```python
from io import BytesIO
from IPython.core.display import Image, display
import PIL
import numpy as np

def display_img(image, fmt='png', retina=True):
    bio = BytesIO()
    image.save(bio, format=fmt)
    display(Image(bio.getvalue(), format=fmt, retina=retina))
```

```python
display_img(PIL.Image.open('./res/table_real.jpg'))
```

```python
from recognize import LetterRecognizer

rec = LetterRecognizer(image_path='./res/table_real.jpg')
```

```python
letters = rec.find_letters()
```

```python
display_img(letters[0])
```

```python
some_letter = letters[0]
```

```python
image_size = (64, 64)
image = PIL.Image.new('L', image_size, 255)

left = int(image.width / 2 - some_letter.width / 2)
top = int(image.height / 2 - some_letter.height / 2)
right = left + some_letter.width
bottom = top + some_letter.height

letter_bounds = (left, top, right, bottom)

image.paste(some_letter, letter_bounds)

display_img(image)
```
